```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_benin=pd.read_csv('Data/benin-malanville.csv')
df_sierraleone=pd.read_csv('Data/sierraleone-bumbuna.csv')
df_togo=pd.read_csv('Data/togo-dapaong_qc.csv')

datas=[df_benin,df_sierraleone,df_togo]
```

```python
# Remove duplicates

for data in datas:
    data.drop_duplicates(inplace=True)
    
```

```python
# Drop column 'Comments'(no use)

for data in datas:
    data.drop("Comments", axis=1, inplace=True)
```

```python
for df in datas:
    df[['GHI', 'DNI', 'DHI']] = df[['GHI', 'DNI', 'DHI']].applymap(lambda x: np.nan if x < 0 else x)

```

```python
for df in datas:
    negative_counts = {}
    for column in df.select_dtypes(include='number'):
        # Count negative values in the column
        negative_counts[column] = (df[column] < 0).sum()

    # Print the counts of negative values for each column
    print("Negative Value Counts:")
    for column, count in negative_counts.items():
        print(f"{column}: {count}")
```

```python
df_benin=pd.read_csv('benin_pp_ds.csv')
df_sierraleone=pd.read_csv('togo_pp_ds.csv')
df_togo=pd.read_csv('sierraleone_pp_ds.csv')

datas=[df_benin,df_sierraleone,df_togo]
```

```python
data=df_benin
numeric_columns = ['GHI', 'DNI', 'DHI', 'Tamb', 'TModA', 'TModB']

# Calculate Z-scores for selected columns
z_scores = (data[numeric_columns] - data[numeric_columns].mean()) / data[numeric_columns].std()

# Define threshold for outlier detection
threshold = 3

# Identify outliers
outliers = data[(np.abs(z_scores) > threshold).any(axis=1)]

# Calculate the total number of observations
total_observations = len(data)
number_of_outliers = len(outliers)

# Calculate the percentage of outliers
percentage_outliers = (number_of_outliers / total_observations) * 100

print("Percentage of outliers:", percentage_outliers)
```

```python
data=df_benin


# Convert relevant columns to numeric data type
data['GHI'] = pd.to_numeric(data['GHI'], errors='coerce')
data['DNI'] = pd.to_numeric(data['DNI'], errors='coerce')
data['DHI'] = pd.to_numeric(data['DHI'], errors='coerce')
data['Tamb'] = pd.to_numeric(data['Tamb'], errors='coerce')
data['TModA'] = pd.to_numeric(data['TModA'], errors='coerce')
data['TModB'] = pd.to_numeric(data['TModB'], errors='coerce')



numeric_data = data.select_dtypes(include=['float64', 'int64'])  # Select only numeric columns
z_scores = (numeric_data - numeric_data.mean()) / numeric_data.std()

# Define threshold for outlier detection (e.g., 3 standard deviations from the mean)
outlier_threshold = 3

# Identify outliers
outliers = (z_scores > outlier_threshold) | (z_scores < -outlier_threshold)

# Remove rows with outliers
data_cleaned = data[~outliers.any(axis=1)]
```

```python
data=data_cleaned
# Filter out -inf values from log-transformed columns
data_no_inf = data.replace([np.inf, -np.inf], np.nan).dropna()

# Plot histograms for all columns in the DataFrame after log transformation
data_no_inf.hist(figsize=(12, 10))
plt.show()
```
